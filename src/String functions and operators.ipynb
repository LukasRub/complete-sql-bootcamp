**_Example:_** Do movies that have shorter titles get rented more often? What is the relationship between the times a movie is rented and it's title lenght?

```python
with db_connection() as conn:
    conn.fetch_all("""
        SELECT COUNT(rental.rental_id) AS times_rented,
               length(film.title) AS title_length
        FROM rental 
        INNER JOIN inventory 
            ON rental.inventory_id = inventory.inventory_id
        INNER JOIN film
            ON inventory.film_id = film.film_id
        GROUP BY title_length
        ORDER BY title_length ASC;
    """)
```
