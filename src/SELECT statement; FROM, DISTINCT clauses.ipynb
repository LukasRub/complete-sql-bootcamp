### **SELECT** fundamentals

### SELECT DISTINCT

```python
with db_connection() as conn:
    conn.fetch_all("SELECT DISTINCT release_year FROM film;")
```

**DISTINCT** works both as a query and as a function.

```python
with db_connection() as conn:
    conn.fetch_all("SELECT DISTINCT(release_year) FROM film;")
```

```python
with db_connection() as conn:
    conn.fetch_all("SELECT DISTINCT(rental_rate) FROM film;")
```
