### **`WHERE`** clause

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT first_name, last_name from actor
                      WHERE first_name = 'Nick';""")
```

With **`AND`** conditional.

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT * from film
                      WHERE rental_rate > 4 
                          AND replacement_cost >= 19.99 
                          AND rating = 'R';""")
```

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT COUNT(title) from film
                      WHERE rental_rate > 4 
                          AND replacement_cost >= 19.99 
                          AND rating = 'R';""")
```

With **`OR`** conditional.

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT COUNT(title) from film
                      WHERE rating = 'R'
                          OR rating = 'PG-13';""")
```

With **`NOT`** conditional.

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT COUNT(title) from film
                      WHERE rating != 'R'
                          OR rating = 'PG-13';""")
```
