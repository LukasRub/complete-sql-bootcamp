### **`ORDER BY`** clause

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT * FROM customer
                      ORDER BY first_name;""")
```

**`ORDER BY`** orders in ascending order by default, but **`ASC`** flag can be explictly included.

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT * FROM customer
                      ORDER BY first_name ASC;""")
```

You can chain multiple columns with which to **`ORDER BY`**, with each having separate order flags (**`ASC`**/**`DESC`**).

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT store_id, first_name, last_name FROM customer
                      ORDER BY store_id DESC, 
                               first_name ASC;""")
```

You can also sort by columns you don't request in your **`SELECT`** statement.

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT first_name, last_name FROM customer
                      ORDER BY store_id DESC, 
                               first_name ASC;""")
```

### **`LIMIT`** clause

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT * FROM payment
                      ORDER BY payment_date DESC
                      LIMIT 10;""")
```

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT * FROM payment
                      WHERE amount > 0
                      ORDER BY payment_date DESC,
                               amount ASC
                      LIMIT 10;""")
```

### **`ORDER BY`**, **`LIMIT`** challange

```python
with db_connection() as conn:
    conn.fetch_all("""SELECT customer_id, amount, payment_date
                      FROM payment
                      ORDER BY payment_date ASC
                      LIMIT 10;""")
```
